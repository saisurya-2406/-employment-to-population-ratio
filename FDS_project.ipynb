  **EMPLOYEMENT TO POLPULATION RATIO**             

The dataset is about the employment-to-population ratio, broken down by gender and age group across countries.
This dataset from the United Nations Statistics Division, titled Employment-to-Population Ratio, is part of the Gender Info 2007 database. It provides percentage values of employed individuals relative to the total population for various countries, segmented by:
- Gender: Male and Female
- Age groups: 15+ years and 15–24 years
- Years: Covers data from 1991 to 2006
- Source: Primarily from the ILO KILM Database (5th edition, 2007)


**Exploratory Data Analysis (EDA)** is the process of examining, understanding, and summarizing datasets before applying any modeling or machine learning techniques.
It helps data analysts and scientists gain insights into data structure, detect anomalies, and prepare the dataset for further analysis.

EDA involves both statistical summaries and data visualizations to identify important patterns, relationships, and trends.

**Objectives of EDA**

*To understand the structure and nature of the dataset.

*To detect missing values, incorrect data types, and duplicates.

*To summarize numerical and categorical features.

*To visualize data distributions and relationships.

*To identify outliers and correlations between variables.

*To prepare cleaned data for further modeling or analysis.

```python
#remove the data in cloumn
data.drop(['Value Footnotes'],axis=1, inplace=True)
data.head()
```

```python
#finding the duplicates
data.duplicated().sum()
```

```python
# identifying the garbage value
for i in data.select_dtypes(include=['object']):
  print(data[i].value_counts())
```

```python
#outliers
numeric_cols = data.select_dtypes(include=np.number).columns
for col in numeric_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = data[(data[col] < (Q1 - 1.5 * IQR)) | (data[col] > (Q3 + 1.5 * IQR))]
    print(f"{col}: {len(outliers)} outliers detected")
```

```python
# relationship analysis
data.select_dtypes(include=['number']).corr()
```

```python
# Plotting a line graph of Value over the years
plt.figure(figsize=(12, 6))
plt.plot(data['Year'], data['Value'])
plt.xlabel('Year')
plt.ylabel('Employment-to-Population Ratio (Percent)')
plt.title('Employment-to-Population Ratio Over Years')
plt.grid(True)
plt.show()
```

```python
average_value_by_year = data.groupby('Year')['Value'].mean().reset_index()
plt.figure(figsize=(10, 6))
plt.plot(average_value_by_year['Year'], average_value_by_year['Value'], marker='o', linestyle='-')
plt.title('Average Value Over Years')
plt.xlabel('Year')
plt.ylabel('Average Value')
plt.grid(True)
plt.show()
```

```python
#histogram
plt.figure(figsize=(10, 6))
plt.hist(data['Value'], bins=30, edgecolor='black')
plt.xlabel('Employment-to-Population Ratio (Percent)')
plt.ylabel('Frequency')
plt.title('Distribution of Employment-to-Population Ratio')
plt.grid(True)
plt.show()
```

```python
#piechart with subgroup
subgroup_counts = data['Subgroup'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(subgroup_counts, labels=subgroup_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Data by Subgroup')
plt.axis('equal')
plt.show()
```

```python
#area graph with year and value
plt.figure(figsize=(12, 6))
plt.fill_between(data['Year'], data['Value'])
plt.xlabel('Year')
plt.ylabel('Employment-to-Population Ratio (Percent)')
plt.title('Employment-to-Population Ratio Over Years (Area Plot)')
plt.grid(True)
plt.show()
```

```python
#average_value_by_yearby area graph
average_value_by_year = data.groupby('Year')['Value'].mean().reset_index()
plt.figure(figsize=(10, 6))
plt.fill_between(average_value_by_year['Year'], average_value_by_year['Value'], color='skyblue', alpha=0.4)
plt.plot(average_value_by_year['Year'], average_value_by_year['Value'], color='Slateblue', alpha=0.6)
plt.title('Average Value Over Years (Area Graph)')
plt.xlabel('Year')
plt.ylabel('Average Value')
plt.grid(True)
plt.show()
```

```python
# boxplot
data_to_plot = [data[data['Year'] == year]['Value'] for year in sorted(data['Year'].unique())]
plt.figure(figsize=(12, 6))
plt.boxplot(data_to_plot, labels=sorted(data['Year'].unique()))
plt.xlabel("Year")
plt.ylabel("Value")
plt.title("Distribution of Value by Year")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
